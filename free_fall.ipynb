# Free Fall
This notebook demonstrates how to construct a trajetory of a massive body subject to free fall with different gravitational constants.

```python
# numpy is a covenient library for fast numerical computation
import numpy as np
# scipy contains an ordinary differential equation (ODE) solver
from scipy.integrate import solve_ivp
# visualization library
import matplotlib.pyplot as plt
```

```python
# define a function that returns the distance from the origin as function of time
def freefall(t,zdot0,z0,g):
    return z0+zdot0*t+0.5*g*t*t
```

```python
# define constants and time
t = np.linspace(0, 7, 300)
gmoon = -1.62
gearth = -9.81
z0 = 20
zdot0 = 0
```

```python
# plot the flight path velocity v
plt.figure(dpi=150)
plt.plot(t, freefall(t,zdot0,z0,gearth),label='Earth')
plt.plot(t, freefall(t,zdot0,z0,gmoon),'--',label='Moon')
plt.xlabel('t [s]')
plt.ylabel('h [m]')
plt.title('Free Fall')
plt.legend()
plt.ylim(0,z0+5)
plt.show()
```

```python
# reuse the same function for different constants 
t = np.linspace(0, 7, 300)
gmoon = -1.62
gearth = -9.81
z0 = 20
zdot0 = 5
```

```python
# plot the flight path velocity v
plt.figure(dpi=150)
plt.plot(t, freefall(t,zdot0,z0,gearth),label='Earth')
plt.plot(t, freefall(t,zdot0,z0,gmoon),'--',label='Moon')
plt.xlabel('t [s]')
plt.ylabel('h [m]')
plt.title('Free Fall')
plt.legend()
plt.ylim(0,z0+2*zdot0)
plt.show()
```
